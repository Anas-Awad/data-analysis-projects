### importing libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as pl
import plotly.express as px
%matplotlib inline
sns.set_theme(context = 'poster' , style = 'ticks' , palette = 'Greys_r' , font_scale = 0.7 , rc = {'figure.figsize':[21 , 9]})
```

### pd.read_csv - To import the CSV file in Jupyter notebook

```python
cars = pd.read_csv('Cars Data1.csv')
cars
```

### head() - It shows the first N rows in the data (by default, N=5)

```python
cars.head()
```

### shape - It shows the total no. of rows and no. of columns of the dataframe

```python
cars.shape
```

### df.isnull( ).sum( ) - It detects the missing values from each column of the dataframe.

```python
cars.isnull().sum()
```

### .info() - To Get info about the data set features types and memory usage and shape of dataset also

```python
cars.info()
```

### Dropna() - to drop all the rows which contains nan values

```python
cars.dropna(inplace=True)
cars
```

### fillna() - To fill the null values of a column with some particular value

```python
mode = cars['Make'].mode()
cars['Make'].fillna(mode)
```

### value_counts - In a column, it shows all the unique values with their count. It can be applied to a single column only.

```python
cars['Make'].value_counts()
```

### isin() - To show all records including particular elements

```python
x = cars['Make'].value_counts().head(10)
```

```python
cars[cars['Make'].isin(x.index)].head(40)
```

### apply() - To apply a function along any axis of DF

```python
cars['MSRP'] = cars['MSRP'].apply(lambda x: x[1:])
cars
```

### Q. 3) Instruction ( Filtering ) - Show all the records where Origin is Asia or Europe.

```python
cars[cars['Origin'].isin(['Asia','Europe'])]
```

### Q. 4) Instruction ( Removing unwanted records ) - Remove all the records (rows) where Weight is above 4000.

```python
indexs = cars[cars['Weight'] > 4000].index
cars.drop(index=indexs,inplace=True)
```

```python
cars
```

### Q. 5) Instruction ( Applying function on a column ) - Increase all the values of 'MPG_City' column by 3

```python
cars['MPG_City'] = cars['MPG_City'].apply(lambda x: x+3)
cars['MPG_City']
```

### Q. 6) What is the most five selling makers

```python
x = cars['Make'].value_counts().head(5)
sns.countplot(data=cars[cars['Make'].isin(x.index)],x = 'Make',order=x.index)
```

### Q. 7) what is the most sold cars brand in europe

```python
europe = cars[cars['Origin'] == 'Europe']
europe
```

```python
most = europe['Make'].value_counts()
most
```

```python
sns.countplot(data=europe,x= 'Make',order=most.index,palette='cool')
```

```python
px.pie(europe,names = 'Make',title= 'Most Sold Cars Brand In Europe',color_discrete_sequence=px.colors.qualitative.Dark2)
```

### Q. 8) which make has a higher horse power motors

```python
x = cars.groupby('Make')['Horsepower'].mean().sort_values().tail()
x
```

```python
sns.barplot(data=cars[cars['Make'].isin(x.index)],x = 'Make', y= 'Horsepower',order=x.index,palette='inferno')
```

### Q. 9) which make has a bigger Engine size

```python
x = cars.groupby('Make')['EngineSize'].mean().sort_values(ascending = False).head()
x
```

```python
sns.barplot(data=cars[cars['Make'].isin(x.index)],x = 'Make', y= 'EngineSize',order=x.index,palette='rocket')
```
